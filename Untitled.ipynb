```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
```

```python
#data collection and processing 
car_data=pd.read_csv('/users/rahul/Desktop/ML/car.csv')
```

```python
#checking the distribution of categorical data 
print(car_data.Fuel_Type.value_counts())
print(car_data.Seller_Type.value_counts())
print(car_data.Transmission.value_counts())
      
```

```python
#encoding "Fuel-Type"Column
car_data.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2}},inplace=True)

#encoding "Seller-Type"Column
car_data.replace({'Seller_Type':{'Dealer':0,'Individual':1}},inplace=True)

#encoding "Transmission-Type"Column
car_data.replace({'Transmission':{'Manual':0,'Automatic':1}},inplace=True)
```

```python
car_data.head()
```

```python
#splitting data in to target data and test data 

x=car_data.drop(['Car_Name','Selling_Price'],axis=1)
y=car_data['Selling_Price']
```

```python
print (x)
```

```python
print(y)
```

```python
#Splitting training and test data 

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
#training the the madel

lin_reg_model=LinearRegression()
```

```python
lin_reg_model.fit(x_train,y_train)
```

```python
print(lin_reg_model)
```

```python
#model evaluation
#prediction on training data 

training_data_prediction=lin_reg_model.predict(x_train)
```

```python
# R square error
error_score=metrics.r2_score(y_train,training_data_prediction)

print("R Square error :",error_score)
```

```python
#visualize the actual prices and predicted prices

plt.scatter(y_train,training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual price vs predicted price:")
plt.show()
```

```python
test_data_prediction=lin_reg_model.predict(x_test)
```

```python
# R square error
error_score=metrics.r2_score(y_test,test_data_prediction)

print("R Square error :",error_score)
```

```python
#visualize the actual prices and predicted prices

plt.scatter(y_test,test_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual price vs predicted price:")
plt.show()
```

```python
#training the the madel Lasso

lass_reg_model=Lasso()
```

```python
lass_reg_model.fit(x_train,y_train)
```

```python
print(lass_reg_model)
```

```python
#model evaluation
#prediction on training data 

training_data_prediction=lass_reg_model.predict(x_train)
```

```python
# R square error
error_score=metrics.r2_score(y_train,training_data_prediction)

print("R Square error :",error_score)
```

```python
#visualize the actual prices and predicted prices

plt.scatter(y_train,training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual price vs predicted price:")
plt.show()
```

```python
test_data_prediction=lass_reg_model.predict(x_test)
```

```python
# R square error
error_score=metrics.r2_score(y_test,test_data_prediction)

print("R Square error :",error_score)
```

```python
#visualize the actual prices and predicted prices

plt.scatter(y_test,test_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual price vs predicted price:")
plt.show()
```
